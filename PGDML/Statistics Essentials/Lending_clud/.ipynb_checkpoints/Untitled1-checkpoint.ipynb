## DATA CLEANING

```python
np.array((loans.isnull().sum()/len(loans.index)*100)>90)
```

### Cleaning Null Columns - By checking null values greater than 90% 

```python
loans=loans.iloc[:,np.array((loans.isnull().sum()/len(loans.index)*100)<=90)]
```

```python
loans.shape
```

```python
loans.info()
```

```python
loans.sample(10)
```

### Converting term to int by removing stirng "months"

```python
loans.term.value_counts()
```

```python
loans.term=loans.term.apply(lambda x: int(x.split(" ")[1]))
```

### Converting int_rate	 to float by removing %

```python
loans.int_rate.value_counts()
```

```python
loans.int_rate=loans.int_rate.apply(lambda x: float(x[:-1]))
```

```python
loans.shape
```

```python
loans.describe(include='all').transpose()
```

### Dropping emp_title column,url, desc, pymnt_plan etc as data is either static or not necesary for analysis

```python
loans=loans.drop('emp_title',axis=1)
loans=loans.drop('pymnt_plan',axis=1)
loans=loans.drop('url',axis=1)
loans=loans.drop('desc',axis=1)
loans=loans.drop('initial_list_status',axis=1)
loans=loans.drop('policy_code',axis=1)
loans=loans.drop('application_type',axis=1)
loans=loans.drop('acc_now_delinq',axis=1)
loans=loans.drop('tax_liens',axis=1)
loans=loans.drop('delinq_amnt',axis=1)
loans=loans.drop('chargeoff_within_12_mths',axis=1)
loans=loans.drop('collections_12_mths_ex_med',axis=1)
loans=loans.drop('title',axis=1)
```

```python
loans.sample(5)
```

### Removing Current Loans from Analysis

```python
loans=loans[~(loans.loan_status=='Current')]
```

```python
loans.shape
```

```python
loans.loan_status.value_counts()
```

```python
loans.isnull().sum()*100/len(loans.index)
```

### Rounding off funded_amnt_inv 2 round decimals

```python
loans.funded_amnt_inv= loans.funded_amnt_inv.apply(lambda x: round(x,2))
```

```python
loans.head()
```

```python
loans.describe(include='all').transpose()
```

### Converting emp_length to int by removing +, > etc

```python
loans.emp_length.value_counts()
```

```python
loans.emp_length.isnull().sum()
```

## Assumption - Since, we have 1033 NAN in emp_length, we will be filling it up with mode which is 10+ years

```python
loans.emp_length=loans.emp_length.fillna(loans.emp_length.mode()[0])
```

```python
loans[loans.emp_length.isnull()]
```

```python
loans.emp_length.isnull().sum()
```

### Cleaning the dataset of emp_length

```python
def clean_emp_length(emp_length):
    x=emp_length.split(" ")
    if(len(x)==3):
        return 0
    else:
        if((x[0].find("+")==-1)):
            return x[0]
        else:
            return x[0][:-1]
```

```python
loans.emp_length=loans.emp_length.apply(lambda x: float(clean_emp_length(x)))
```

```python
loans.head()
```

```python
### Creating Issue Year from loan Issue_d
```

```python
loans['issue_y']=loans.issue_d.apply(lambda x: x.split("-")[1])
```

```python
loans['dti_bucket']=pd.cut(loans.dti,bins=[0,10,20,loans.dti.max()],labels=["L Dti","Moderate Dti","H Dti"])
```
